```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
```

```python
data = pd.read_csv('files_for_lab/customer_churn.csv')
data
```

```python
data.shape
```

```python
data.isnull().sum()
```

```python
data.dtypes
```

```python
data['Churn'] = np.where(data['Churn'] == 'No', 0, 1)
```

```python
data
```

# Assumptions of Logistic Reg

```python
y = data['Churn']
X = data[['SeniorCitizen', 'tenure', 'MonthlyCharges']]
X
```

```python
correlations_matrix = X.corr()
sns.heatmap(correlations_matrix, annot=True)
plt.show()
```

## Standardization of independent variables

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
```

```python
transformer = StandardScaler().fit(X_train)
X_train_scaled = pd.DataFrame(transformer.transform(X_train),columns=X.columns)
X_test_scaled = pd.DataFrame(transformer.transform(X_test),columns=X.columns)
X_train_scaled.head()
```

```python
X_train_scaled
```

```python
y_train.head()
```

```python
y_train = y_train.reset_index(drop=True)
# needed because the transformation also reset the index for X_train_scaled
# so we need this for y_test as well...
y_test = y_test.reset_index(drop=True)
y_train
```

```python
LR = LogisticRegression(random_state=0, solver='lbfgs')
LR.fit(X_train_scaled, y_train)
LR.score(X_test_scaled, y_test)
```

```python
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score

pred = LR.predict(X_test_scaled)

print("precision: ",precision_score(y_test,pred))
print("recall: ",recall_score(y_test,pred))
print("f1: ",f1_score(y_test,pred))
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,pred)
```

```python
count_classes = y_train.value_counts()
count_classes
count_classes.plot(kind = 'bar')
```

```python
#!pip install imbalanced-learn
# this used to be !pip install imblearn
from imblearn.over_sampling import SMOTE
```

```python
sm = SMOTE(random_state=100, k_neighbors=3)
X_train_SMOTE,y_train_SMOTE = sm.fit_resample(X_train_scaled,y_train)
```

```python
from sklearn.linear_model import LogisticRegression

LR = LogisticRegression(random_state=0, solver='lbfgs')
LR.fit(X_train_SMOTE, y_train_SMOTE)
pred = LR.predict(X_test_scaled)

print("precision: ",precision_score(y_test,pred))
print("recall: ",recall_score(y_test,pred))
print("f1: ",f1_score(y_test,pred))
```

```python
count_classes = y_train_SMOTE.value_counts()
count_classes
count_classes.plot(kind = 'bar')
```
